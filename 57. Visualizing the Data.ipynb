## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here


QUERY = """
SELECT * FROM DatabaseDesireNextYear
"""

df2 = pd.read_sql_query(QUERY,conn)
print(df2.shape)
df2.head()
```

```python
df2.duplicated().sum() # no duplicated rows, each respondent only responded with unique databases
```

```python
df2["DatabaseDesireNextYear"].value_counts(ascending=False).head() # top 5 databases
```

```python
df2[df2["DatabaseDesireNextYear"] == "MySQL"].count() # example result for the for loop reference
```

```python
# get as list
to_learn_top5 = df2["DatabaseDesireNextYear"].value_counts(ascending=False).head().index.to_list()
number_of_respondents = []  # empty list for the result set

for database in to_learn_top5:
    df_loop = df2[df2["DatabaseDesireNextYear"] == database]
    number_of_respondents.append(df_loop.count()[1])  # retrive the second element in the result series

print(number_of_respondents)
```

```python
# define Seaborn color palette to use
palette_color = sns.color_palette('bright')

# declaring exploding pie for highest value
explode = [0.1, 0, 0, 0, 0]

plt.pie(number_of_respondents, labels=to_learn_top5, colors=palette_color, autopct='%.0f%%', explode=explode)
plt.rcParams["figure.figsize"] = (15,15)
plt.title("Top 5 Databases Respondents Want to Learn")
# displaying chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
WHERE Age BETWEEN 30 AND 35
"""

df3 = pd.read_sql_query(QUERY,conn)
df3.head()
```

```python
age_range = [30,31,32,33,34,35]
WorkWeekHrs_Medians = []
CodeRevHrs_Medians = []
for age in age_range:
    median1 = df3[df3["Age"] == age]["WorkWeekHrs"].median()
    WorkWeekHrs_Medians.append(median1)
    median2 = df3[df3["Age"] == age]["CodeRevHrs"].median()
    CodeRevHrs_Medians.append(median2)

print(WorkWeekHrs_Medians)
print(CodeRevHrs_Medians)
```

```python
plt.bar(age_range, WorkWeekHrs_Medians, color='r')
plt.bar(age_range, CodeRevHrs_Medians, bottom=WorkWeekHrs_Medians, color='b')

plt.xlabel("Age")
plt.ylabel("Median Hours")
plt.ylim(0,50)
plt.legend(["Work Week Hours", "Code Revision Hours"])
plt.title("Median Work Week Hours & Code Revision Hours By Age")
plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp FROM master
WHERE Age BETWEEN 45 AND 60
"""

df4 = pd.read_sql_query(QUERY,conn)
df4.head()
```

```python
age_range = [*range(45,61,1)]  # ages 45 to 60
Annual_Salary_Medians = []

for age in age_range:
    median = df4[df4["Age"] == age]["ConvertedComp"].median()
    Annual_Salary_Medians.append(median)

print(Annual_Salary_Medians)
```

```python
plt.plot(age_range, Annual_Salary_Medians)

plt.xlabel("Age")
plt.ylabel("Annual Salary")
plt.ylim(80000,160000)
plt.yticks(np.arange(80000,160001, step=20000))
plt.xlim(45,61)
plt.xticks(np.arange(45,61, step=5))
plt.title("Median Annual Salary VS Age")
plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```
